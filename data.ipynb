About Dataset
This data set contains 13 clinicopathologic features aiming to predict recurrence of well differentiated thyroid cancer. The data set was collected in duration of 15 years and each patient was followed for at least 10 years.

```python
db["Physical Examination"].unique()
```

```python
db["Physical Examination"] = db[["Physical Examination"]].replace({"Single nodular goiter-left":0,"Multinodular goiter":1,"Single nodular goiter-right":2,"Normal', 'Diffuse goiter":3})
db.head()
```

```python
db["Adenopathy"].unique()
```

```python
db["Adenopathy"] = db[["Adenopathy"]].replace({"No":0,"Right":1,"Extensive":2,"Left":3,"Bilateral":4,"Posterior":5})
db.head()
```

```python
db["Recurred"].unique()
```

```python
db["Recurred"] = db[["Recurred"]].replace({"No":0,"Yes":1})
db.head()
```

```python
db["Stage"].unique()
```

```python
db["Stage"] = db[["Stage"]].replace({"I":1,"II":2,"IVB":3,"III":4,"IVA":5})
db.head()
```

```python
db["Response"].unique()
```

```python
db["Response"] = db[["Response"]].replace({"Indeterminate":0,"Excellent":1,"Structural Incomplete":2,"Biochemical Incomplete":3})
db.head()
```

```python
db["Thyroid Function"].unique()
```

```python
db["Thyroid Function"] = db[["Thyroid Function"]].replace({"Euthyroid":0,"Clinical Hyperthyroidism":1,"Clinical Hypothyroidism":2,"Subclinical Hyperthyroidism":3,"Subclinical Hypothyroidism":4})
db.head()
```

```python
db["Pathology"].unique()
```

```python
db["Pathology"] = db[["Pathology"]].replace({"Micropapillary":0,"Papillary":1,"Follicular":2,"Hurthel cell":"3"})
db.head()
```

```python
db["Risk"].unique()
```

```python
db["Risk"] = db[["Risk"]].replace({"Low":0,"Intermediate":1,"High":2})
db.head()
```

```python
db.drop(db[["Focality","T","N","M"]],axis = 1,inplace=True)
```
